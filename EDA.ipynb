## EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "Cleaned_Sorted_SummerStudentAdmissions.csv"
df = pd.read_csv(file_path)

# Display basic information
print(df.info())

# Summary statistics for numerical columns
print(df.describe())

# Set style for plots
sns.set_style("whitegrid")

# GPA Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df["GPA"], bins=15, kde=True)
plt.title("GPA Distribution")
plt.xlabel("GPA")
plt.ylabel("Frequency")
plt.show()

# Work Experience Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df["WorkExp"], bins=15, kde=True)
plt.title("Work Experience Distribution")
plt.xlabel("Work Experience (Years)")
plt.ylabel("Frequency")
plt.show()

# Test Score Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df["TestScore"], bins=15, kde=True)
plt.title("Test Score Distribution")
plt.xlabel("Test Score")
plt.ylabel("Frequency")
plt.show()

# Writing Score Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df["WritingScore"], bins=15, kde=True)
plt.title("Writing Score Distribution")
plt.xlabel("Writing Score")
plt.ylabel("Frequency")
plt.show()

# Volunteer Level Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x=df["VolunteerLevel"], palette="viridis")
plt.title("Volunteer Level Distribution")
plt.xlabel("Volunteer Level")
plt.ylabel("Count")
plt.show()

# Total Score Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df["TotalScore"], bins=15, kde=True)
plt.title("Total Score Distribution")
plt.xlabel("Total Score")
plt.ylabel("Frequency")
plt.show()

# Admission Decision Count
plt.figure(figsize=(8, 6))
sns.countplot(x=df["Decision"], palette="coolwarm")
plt.title("Admission Decision Count")
plt.xlabel("Decision")
plt.ylabel("Count")
plt.show()

# Gender Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x=df["Gender"], palette="pastel")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# Experience Level Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x=df["ExperienceLevel"], palette="magma")
plt.title("Experience Level Distribution")
plt.xlabel("Experience Level")
plt.ylabel("Count")
plt.show()

# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

# Boxplot of GPA by Decision
plt.figure(figsize=(8, 6))
sns.boxplot(x=df["Decision"], y=df["GPA"], palette="Set2")
plt.title("GPA Distribution by Admission Decision")
plt.xlabel("Decision")
plt.ylabel("GPA")
plt.show()

# Boxplot of Total Score by Decision
plt.figure(figsize=(8, 6))
sns.boxplot(x=df["Decision"], y=df["TotalScore"], palette="Set3")
plt.title("Total Score Distribution by Admission Decision")
plt.xlabel("Decision")
plt.ylabel("Total Score")
plt.show()

# Boxplot of Work Experience by Decision
plt.figure(figsize=(8, 6))
sns.boxplot(x=df["Decision"], y=df["WorkExp"], palette="Blues")
plt.title("Work Experience Distribution by Admission Decision")
plt.xlabel("Decision")
plt.ylabel("Work Experience (Years)")
plt.show()
```
